```python
# import packages
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/electric_vehicles.csv", sep=";", index_col=0)
df['City'] = df['City'].str.replace('\d+ ', '')
```

```python
dv_ev_final = pd.read_excel("../data/excel.xlsx")
dv_ev_final
```

```python
df = pd.merge(df, dv_ev_final, on="City")
df
```

```python
growth_rate_2021=(df["2021"]-df["2020"])/df["2020"]
growth_rate_2020=(df["2020"]-df["2019"])/df["2019"]
growth_rate_2019=(df["2019"]-df["2018"])/df["2018"]
growth_rate_2018=(df["2018"]-df["2017"])/df["2017"]
gr_mean=(growth_rate_2021+growth_rate_2020+growth_rate_2019+growth_rate_2018)/4
df["2022_gr"] = round(df["2021"]*(1+gr_mean))
df["recomm_2022"] = round(df["Count"]*(1+(df["2022_gr"]-df["2021"])/df["2021"]))
df["diff_2022"] = df["recomm_2022"]-df["Count"]
df["2023_gr"] = round(df["2021"]*((1+gr_mean)**2))
df["recomm_2023"] = round(df["recomm_2022"]*(1+(df["2023_gr"]-df["2022_gr"])/df["2022_gr"]))
df["diff_2023"] = df["recomm_2023"]-df["recomm_2022"]
df["2024_gr"] = round(df["2021"]*((1+gr_mean)**3))
df["recomm_2024"] = round(df["recomm_2023"]*(1+(df["2024_gr"]-df["2023_gr"])/df["2023_gr"]))
df["diff_2024"] = df["recomm_2024"]-df["recomm_2023"]
df
```

```python
df2 = df.dropna()
df2
```
