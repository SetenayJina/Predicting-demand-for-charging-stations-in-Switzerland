```python
# import packages
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as st
from matplotlib import rcParams
```

```python
df = pd.read_csv("../data/electric_vehicles.csv", sep=";", index_col=0)
df['City'] = df['City'].str.replace('\d+ ', '')
df
```

```python
dv_ev_final = pd.read_excel("../data/excel.xlsx")
dv_ev_final
```

```python
df = pd.merge(df, dv_ev_final, on="City")
df = df.dropna()
df
```

```python
y = df['Count']
x = df.drop(columns = ['Count',"City"])

# with GLM (generalized linear model)
my_model = sm.OLS(y, sm.add_constant(x)).fit()

print(my_model.summary())
```

```python
dfz = df.loc[df["City"] == "Zürich"]
dfz
```

```python
y2 = dfz['Count']
x2 = dfz.drop(columns = ['Count',"City"])

# with GLM (generalized linear model)
my_model = sm.OLS(y2, sm.add_constant(x2)).fit()

print(my_model.summary())
```
